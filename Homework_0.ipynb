## Loading, viewing, and cropping a grayscale image:

```python
# import sys
import cv2 as cv
import matplotlib.pyplot as plt
import os
```

```python
# check that the file exists & load it
img_path = 'Data/Apollo_11_Launch.jpg'
assert os.path.exists(img_path), "Cannot find image, check the provided image path is correct."
img = cv.imread(img_path)

# View the image in the cell
plt.imshow(img)
plt.show()
```

```python
# cropping the image
cropped_img = img[5:625, 500:700]
plt.imshow(cropped_img)
plt.show()
```

## Editing the Brightness, Rotating & Saving

```python
# Editing the brightness (using example from class)
import numpy as np
brightening_factor = np.ones(cropped_img.shape, dtype="uint8") * 25
fig, axes = plt.subplots(1, 3)

# brighter
brighter = cv.add(cropped_img, brightening_factor)

# darker
darker = cv.subtract(cropped_img, brightening_factor)

# plot
axes[0].imshow(darker)
axes[0].set_title('Darker')
axes[1].imshow(cropped_img)
axes[1].set_title('Original')
axes[2].imshow(brighter)
axes[2].set_title('Brighter')
plt.show()
```

```python
# Rotating
rotated_img = cv.rotate(cropped_img, cv.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotated_img)
plt.show()
```

```python
# Save image
assert cv.imwrite('Data/submission.jpg', rotated_img), "Could not save image"
```
